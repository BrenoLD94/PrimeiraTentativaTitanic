```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
file = "../Treino/train.csv"
train_data = pd.read_csv(file, index_col=0)

file2 = "../Teste/test.csv"
test_data = pd.read_csv(file2, index_col=0)
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data.drop(['Name', 'Cabin', 'Ticket'],axis=1, inplace=True)
test_data.drop(['Name', 'Cabin', 'Ticket'],axis=1, inplace=True)
```

```python
train_data['Family'] = train_data['SibSp'] + train_data['Parch']
train_data.drop(['SibSp', 'Parch'], inplace=True, axis=1)

test_data['Family'] = test_data['SibSp'] + test_data['Parch']
test_data.drop(['SibSp', 'Parch'], inplace=True, axis=1)
```

```python
def sex(s):
    if(s == 'male'):
        return 1
    elif(s == 'female'):
        return 0
    
train_data['Sex'] = train_data['Sex'].apply(sex)
test_data['Sex'] = test_data['Sex'].apply(sex)
```
